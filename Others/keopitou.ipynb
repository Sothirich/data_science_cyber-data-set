```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Data/Data_of_Attack_Back.csv')
df2=pd.read_csv('Data/Data_of_Attack_Back_BufferOverflow.csv')
df3=pd.read_csv('Data/Data_of_Attack_Back_FTPWrite.csv')
df4=pd.read_csv('Data/Data_of_Attack_Back_GuessPassword.csv')
df5=pd.read_csv('Data/Data_of_Attack_Back_NMap.csv')
df6=pd.read_csv('Data/Data_of_Attack_Back_Neptune.csv')
df7=pd.read_csv('Data/Data_of_Attack_Back_Normal.csv')
df8=pd.read_csv('Data/Data_of_Attack_Back_PortSweep.csv')
df9=pd.read_csv('Data/Data_of_Attack_Back_RootKit.csv')
df10=pd.read_csv('Data/Data_of_Attack_Back_Satan.csv')
df11=pd.read_csv('Data/Data_of_Attack_Back_Smurf.csv')
```

```python
# check it row and column make sure the data are match

print('df1',df1.shape)
print('df2',df2.shape)
print('df3',df3.shape)
print('df4',df4.shape)
print('df5',df5.shape)
print('df6',df6.shape)
print('df7',df7.shape)
print('df8',df8.shape)
print('df9',df9.shape)
print('df10',df10.shape)
print('df11',df11.shape)
```

```python
# add missing header to Data_of_Attack_Back_FTPWrite.csv

header = df1.columns
df3.columns = header
```

```python
# add new column attack 

df1['attack']=['data_of_attack_back']*df1.shape[0]
df2['attack']=['data_of_attack_back_BufferOverflow']*df2.shape[0]
df3['attack']=['data_of_attack_back_FTPWrite']*df3.shape[0]
df4['attack']=['data_of_attack_back_GuessPassword']*df4.shape[0]
df5['attack']=['data_of_attack_back_NMap']*df5.shape[0]
df6['attack']=['data_of_attack_back_Neptune']*df6.shape[0]
df7['attack']=['data_of_attack_back_Normal']*df7.shape[0]
df8['attack']=['data_of_attack_back_PortSweep']*df8.shape[0]
df9['attack']=['data_of_attack_back_RootKit']*df9.shape[0]
df10['attack']=['data_of_attack_back_Satan']*df10.shape[0]
df11['attack']=['data_of_attack_back_Smurf']*df11.shape[0]
```

```python
# concate all the data to one file

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11])
df.reset_index()

# export data
df.to_csv('Data/Data_of_Attack_Appends.csv', index=False)

df = pd.read_csv('Data/Data_of_Attack_Appends.csv')

# clean Unnamed columns
df.columns[0]
# df=df.drop(columns='Unnamed: 0.1',axis='columns')
# df=df.drop(columns='Unnamed: 0',axis='columns')
```

```python
# Pitou

## From df, select only data with attack = 'data_of_attack_back_Normal'
data_attack_back_Normal_df = df[df['attack'] == 'data_of_attack_back_Normal']


## Storing all the unique values of each column in a dictionary
data_attack_back_Normal_unique = {}
for i in data_attack_back_Normal_df.columns:
    data_attack_back_Normal_unique[i] = data_attack_back_Normal_df[i].unique()

## Displaying the unique values of each column
for i in data_attack_back_Normal_unique.keys():
    print(i, ': ', data_attack_back_Normal_unique[i])
    print()
```

```python
# Pitou

## Plotting the data of 'data_of_attack_back_Normal' with x-axis as 'duration' and y-axis as 'src_bytes' and Plotting the data of 'data_of_attack_back_Normal' with x-axis as 'duration' and y-axis as 'dst_bytes' using scatter plot by showing these two visuals next to each other

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].scatter(data_attack_back_Normal_df['duration'], data_attack_back_Normal_df['src_bytes'])
ax[0].set_xlabel('duration')
ax[0].set_ylabel('src_bytes')
ax[0].set_title('data_of_attack_back_Normal')
ax[1].scatter(data_attack_back_Normal_df['duration'], data_attack_back_Normal_df['dst_bytes'], color='orange')
ax[1].set_xlabel('duration')
ax[1].set_ylabel('dst_bytes')
ax[1].set_title('data_of_attack_back_Normal')
plt.show()
```
